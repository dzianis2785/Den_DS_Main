6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*).

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа.

```python
# текст запроса
query_6_6 = f'''

SELECT
    experience,
    round(AVG(coalesce((salary_from + salary_to) / 2, salary_from, salary_to)))

FROM
    vacancies

WHERE
    ( name ILIKE '%data scientist%' OR
      name ILIKE '%data science%' OR
      name ILIKE '%исследователь данных%' OR
      name ILIKE '%machine learning%' OR
      name ILIKE '%машинн%обучен%' OR
      name LIKE '%ML%' AND name NOT ILIKE '%HTML%' )
      AND
    ( (salary_from IS NOT NULL AND salary_to IS NOT NULL) OR
      (salary_from IS NULL AND salary_to IS NOT NULL) OR
      (salary_from IS NOT NULL AND salary_to IS NULL) )

GROUP BY
    experience

'''
```

```python
# результат запроса
df = pd.read_sql_query(query_6_6, connection)
print(f'Средняя зарплата дата-сайентиста с опытом работы от 3 до 6 лет : {df.iloc[2,1]}')
```

```python
import plotly # Подключим графическую библиотеку plotly express i pandas
import plotly.express as px
import pandas as pd
```

```python
# текст запроса
query_7_1 = f'''

SELECT
    experience,
    count(id)

FROM
    vacancies

GROUP BY
    experience

ORDER BY
    count(id) DESC

'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_1, connection)
fig = px.pie(df, values='count', names='experience', height= 500, width= 700, title='Круговая диаграмма вакансий по уровню опыта работы')
fig.show()
```

```python
# текст запроса
query_7_2 = f'''

SELECT
    experience,
    round(AVG(coalesce((salary_from + salary_to) / 2, salary_from, salary_to))) as salary_avg

FROM
    vacancies

WHERE
    (salary_from IS NOT NULL AND salary_to IS NOT NULL) OR
    (salary_from IS NULL AND salary_to IS NOT NULL) OR
    (salary_from IS NOT NULL AND salary_to IS NULL)

GROUP BY
    experience

ORDER BY
    salary_avg DESC

'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_2, connection)
display(df)
```

```python
# текст запроса
query_7_5 = f'''

SELECT
    COUNT(v.id)

FROM

    vacancies v
    LEFT JOIN employers_industries ei ON v.employer_id = ei.employer_id
    LEFT JOIN industries i ON ei.industry_id = i.id

WHERE
    i.name = 'Разработка программного обеспечения'

'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_5, connection)
print(f'Количество вакансий у которых в сфере деятельности указано "Разработка программного обеспечения" : {df.iloc[0,0]}')
```

```python
# текст запроса
query_7_6 = f'''

SELECT
    a.name,
    count(a.name) cnt

FROM
    vacancies v
    LEFT JOIN areas a ON v.area_id = a.id

 WHERE
    (v.name ILIKE '%data scientist%') OR
    (v.name ILIKE '%data science%') OR
    (v.name ILIKE '%исследователь данных%') OR
    (v.name ILIKE '%machine learning%') OR
    (v.name ILIKE '%машинн%обучен%') OR
    ( (v.name ILIKE '%ML%') AND (v.name NOT ILIKE '%HTML%') )

GROUP BY
    a.name

ORDER BY
    cnt DESC

LIMIT 5

'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_6, connection)
display(df)# выводим на экран топ-5 городов по количеству data science вакансий
```

```python
# текст запроса
query_7_7 = f'''

SELECT
    e.name,
    count(e.name) cnt

FROM
    vacancies v
    LEFT JOIN employers e ON e.id = v.employer_id

 WHERE
    (v.name ILIKE '%data scientist%') OR
    (v.name ILIKE '%data science%') OR
    (v.name ILIKE '%исследователь данных%') OR
    (v.name ILIKE '%machine learning%') OR
    (v.name ILIKE '%машинн%обучен%') OR
    ( (v.name ILIKE '%ML%') AND (v.name NOT ILIKE '%HTML%') )

GROUP BY
    e.name

ORDER BY
    cnt DESC

LIMIT 5

'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_7, connection)
# выводим на экран топ-5 компаний по количеству data science вакансий
display(df)
```
