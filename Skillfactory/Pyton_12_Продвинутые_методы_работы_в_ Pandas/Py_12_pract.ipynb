```python
import pandas as pd
ratings_movies = pd.read_csv('data/movies_data/ratings_movies.csv')
```

```python
ratings_movies.copy()
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None

ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
```

```python
orders_products = new_orders.merge(
    products,
    how='outer',
    on='Product_ID'
    )
display(orders_products)
```

```python
filtered = orders_products[(orders_products['Статус'] == 'Оплачен, формируется к отправке')|(orders_products['Статус'] == 'Выполнен')]
display(filtered)
```

```python
filtered['amount'] = filtered['Price']*filtered['Количество']
display(filtered)
```

```python
filter_result = filtered.groupby('ID Покупателя')['amount'].sum()
display(filter_result)
```
