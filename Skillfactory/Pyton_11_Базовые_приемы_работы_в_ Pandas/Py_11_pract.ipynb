```python
citibike_tripdata['age'] = 2018 - citibike_tripdata['birth year']
print(citibike_tripdata[citibike_tripdata['age'] > 60].shape[0])
```

```python
citibike_tripdata['stoptime'] = pd.to_datetime(citibike_tripdata['stoptime'], dayfirst=False)
citibike_tripdata['starttime'] = pd.to_datetime(citibike_tripdata['starttime'], dayfirst=False)
```

```python
citibike_tripdata['trip duration'] = citibike_tripdata['stoptime'] - citibike_tripdata['starttime']
display(citibike_tripdata)
```

```python
citibike_tripdata.info()
```

```python
weekend = citibike_tripdata['starttime'].dt.dayofweek
citibike_tripdata['weekend'] = weekend.apply(lambda x: 0 if x < 5 else 1)
display(citibike_tripdata['weekend'].value_counts(normalize=True))
```

```python
time_of_day = citibike_tripdata['starttime'].dt.hour
def partofday(hour):
    if hour in range(0,7):
        return 'night'
    elif hour in range(7,13):
        return 'morning'
    elif hour in range(13,19):
        return 'day'
    elif hour in range(19,24):
        return 'evening'
citibike_tripdata['time_of_day'] = time_of_day.apply(partofday)
display(citibike_tripdata['time_of_day'].value_counts())
```

```python
print((citibike_tripdata[citibike_tripdata['time_of_day'] == 'day'].shape[0])/((citibike_tripdata[citibike_tripdata['time_of_day'] == 'night'].shape[0])))
```
